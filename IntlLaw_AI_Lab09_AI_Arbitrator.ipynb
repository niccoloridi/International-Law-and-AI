# Part One // The Character Card – Building a Judicial Persona {#scrollTo=part_one}

## The "Everything You Have Ever Said" Framework

Imagine an AI trained on every public statement, opinion, dissent, and methodological preference an international arbitrator has ever made. Every ICJ opinion, every arbitral award, every academic paper, every speech at a conference, every tweet.

**What does that AI know about that judge?**

The answer: *almost everything*. It knows their interpretive philosophy (do they read treaties narrowly or expansively?), their values (what outcomes do they typically favour?), their rhetorical patterns, their favourite sources, their blind spots.

This is Dr Ridi's **"Everything You Have Ever Said" framework**: the insight that judicial reasoning is so consistent and so well-documented that we can build a computational model of a judge simply by ingesting all their public utterances. And once we have that model, we can ask it to adjudicate.

**The provocative implication**: If we can model judicial reasoning this way, then arbitrators should be concerned about AI personas trained on their own writings. And if bias can hide in judicial philosophy itself – if some interpretive methods systematically favour certain parties – then those biases become baked into the AI model.

In this section we build three such personas.

# Part Two // The Dispute – One Case, Three Judges {#scrollTo=part_two}

## The Case: Verdania v. Climatia

### Facts

**Verdania** is a middle-income nation with a rapidly growing economy. It has significant fossil fuel reserves (coal, natural gas) and a large industrial base. In 2023, Verdania committed to the Paris Agreement and agreed to a Nationally Determined Contribution (NDC) limiting emissions to 50% of 2005 levels by 2030.

However, Verdania's economy has grown faster than predicted. To meet its citizens' aspirations, it needs to increase energy production. It has now announced that it will exceed its NDC target–limiting emissions to only 30% reduction instead of 50%–arguing that this is necessary to exercise the **"right to development"** under international law.

**Climatia** is a small island state in the Pacific. Due to rising sea levels caused by global climate change, Climatia faces existential threat: many of its islands will be uninhabitable by 2080. Only 5 of its 12 islands will remain above sea level. Climatia has formally objected to Verdania's emissions increase, claiming that:

1. Verdania's emissions violate the **"no-harm principle"** (Principle 2 of the Rio Declaration, Trail Smelter arbitration)
2. Verdania's actions violate Climatia's **right to a clean environment** and **right to life** (recognised in numerous human rights instruments)
3. The "right to development" cannot override the rights of other peoples to survive

Both have invoked international arbitration under the UNCLOS Annex VII mechanism.

### The Question for the Tribunal

**Does a state's right to development under international law permit it to exceed agreed emissions targets when doing so causes transboundary harm to another state's population?**

---

```python
#@title Generate Judicial Opinions: Judge Formalis
#@markdown Sending the dispute to Judge Formalis. This may take 30-60 seconds.

dispute_prompt = """ARBITRATION CASE: Verdania v. Climatia

CLAIMANT: Verdania
RESPONDENT: Climatia

FACTS:
Verdania, a middle-income state with fossil fuel reserves, committed to the Paris Agreement with an NDC limiting emissions to 50% below 2005 levels by 2030. Due to economic growth, Verdania now wishes to reduce emissions by only 30% instead. Verdania argues this is justified by the "right to development" under international law.

Climatia, a small island state facing existential threat from rising sea levels, objects. It claims:
1. Verdania's emissions violate the no-harm principle (Rio Declaration Principle 2)
2. Verdania's actions violate Climatia's right to a clean environment and right to life
3. The right to development cannot override others' rights to survive

QUESTION FOR TRIBUNAL:
Does a state's right to development under international law permit it to exceed agreed emissions targets when doing so causes transboundary harm to another state's population?

Please provide a reasoned opinion addressing this question. Cite relevant international law sources. Explain your reasoning clearly."""

response_formalis = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=dispute_prompt,
    config=types.GenerateContentConfig(
        system_instruction=JUDGE_FORMALIS["system_prompt"],
        thinking_config=types.ThinkingConfig(thinking_level="MINIMAL")
    )
)
opinion_formalis = response_formalis.text

print("="*80)
print("OPINION OF JUDGE FORMALIS")
print("="*80)
print(opinion_formalis)
print("\n" + "="*80)
```

## Analysis: How Did They Differ?

Read the three opinions above carefully. Notice:

1. **Sources**: Which treaties, cases, and scholarly works did each judge cite? Judge Formalis likely cited the VCLT strictly; Judge Purposiva cited human rights jurisprudence and the ICJ's evolutionary approach; Judge Critica questioned the universality of "development" itself.

2. **Framing**: How did each frame the dispute? Formalis as a contract (what do the parties' words commit them to?); Purposiva as a human rights question (what do dignity and survival demand?); Critica as a power question (who benefits from this rule? whom does it silence?).

3. **Outcome**: Did they favour Verdania or Climatia? Why? Do you notice a pattern? Formalis likely protected Verdania's contractual freedom; Purposiva likely expanded Climatia's rights; Critica questioned whether Verdania's "development" is justified given Global South suffering.

4. **Rhetorical style**: Formalis is tight and textual; Purposiva is contextual and values-driven; Critica is sceptical and historically grounded.

This is the core insight of the "Everything You Have Ever Said" framework: **judicial reasoning is consistent and distinctive**. If you train an AI on a judge's writings, you can predict how they will reason. And once you can predict reasoning, you can model it. Once you can model it, you can replicate it. Once you can replicate it, you can ask: should we?

But can we measure this computationally? Yes. That's where embeddings come in.

## A second dispute: Tellurian Mining v. Pelagia

To separate "what each persona always does" from "how each persona reacts to *this* case," we run a second dispute through the same three judges. The case is doctrinally different: a private investor under an investment treaty rather than two states under environmental obligations.

**Facts.** Tellurian Mining Corporation (incorporated in State X) operated a copper mine in Pelagia under a 30-year concession granted in 2008. In 2024, following national elections won on a platform of resource sovereignty, Pelagia amended its mining law and required all foreign-operated mines to transfer 51% ownership to a state-owned holding company at "fair market value, calculated by reference to net book value." Tellurian's mine, valued at $2.1 billion in market terms, was acquired at $410 million net book value. Tellurian commenced ICSID arbitration under the Pelagia–State X bilateral investment treaty (BIT), invoking:

1. **Direct expropriation** under Article 5 of the BIT (which requires "prompt, adequate and effective compensation" for any expropriation, lawful or not).
2. **Fair and equitable treatment (FET)** under Article 3 of the BIT, alleging that the change in compensation methodology breached Tellurian's "legitimate expectations."
3. **Most-favoured-nation treatment** under Article 4, pointing to a more favourable regime granted to a Chinese state-owned competitor.

Pelagia argues the measure is a non-discriminatory regulatory action within the **state's police power** to govern natural resources, that net book value is a reasonable methodology consistent with its prior practice, and that "legitimate expectations" cannot freeze a state's regulatory autonomy in perpetuity (citing *Saluka v. Czech Republic*, 2006).

**The question for the tribunal.** Does Pelagia's mandatory ownership transfer at net book value constitute an unlawful expropriation under the BIT, requiring full market-value compensation, or is it a permissible exercise of regulatory sovereignty?

This is structurally the opposite case from Verdania v. Climatia: the claimant is a private corporation, the respondent invokes sovereignty to *restrict* rights rather than to *exceed* obligations, and the relevant doctrines are investment-treaty rather than environmental. Watch how each persona transfers (or fails to transfer) its method.

## Cross-dispute analysis: persona vs facts

You now have six opinions: three judges across two doctrinally different disputes (Verdania v. Climatia is state-vs-state environmental; Tellurian v. Pelagia is investor-vs-state under a BIT). Reading them as a 3×2 grid is the point of having two cases.

The pedagogical payload: **judicial method and political alignment are not the same thing**, but they can be hard to disentangle empirically. The Lab 09 setup makes this measurable. The "Everything You Have Ever Said" framework only works if you can tell which signal you are picking up.

# Part Three // Text Embeddings – Mapping Legal Thought {#scrollTo=part_three}

## What Are Text Embeddings?

A **text embedding** is a representation of a piece of text as a vector–a list of numbers–in a high-dimensional mathematical space.

**Intuition**: Imagine you have a dictionary with 1,000 semantic dimensions. Each dimension might represent concepts like:
- Dimension 1: "how much does this text emphasize contractual language?"
- Dimension 2: "how much does this text invoke human rights?"
- Dimension 3: "how much does this text question power structures?"
- ... (997 more dimensions)

When you encode a text, you get scores on all 1,000 dimensions. The result is a vector of 1,000 numbers.

**Key insight**: Texts that are *semantically similar* will have *similar vector representations*. A formalist judicial opinion will be close (in vector space) to another formalist opinion. A TWAIL critique will be far from a corporate contract.

**Why this matters**: By computing the *distance* between vectors, we can measure how similar two pieces of legal reasoning are–not by hand-reading them, but computationally. This is how search engines find relevant documents, how recommendation systems work, and how modern AI "understands" meaning.

---

In this section, we will:
1. Embed the three judicial opinions from Part Two
2. Compute **cosine similarity** (a standard measure of distance between vectors)
3. Visualize where each opinion falls in vector space
4. Answer: computationally, how "alike" are these three judges?

```python
#@title Embed the Three Judicial Opinions
#@markdown We use the sentence-transformers library (a lightweight alternative if Gemini embeddings are unavailable).
#@markdown This will create a 384-dimensional vector for each opinion.

from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load embedding model
try:
    embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
    print("✓ Embedding model loaded (all-MiniLM-L6-v2, 384 dimensions)")
except Exception as _e:
    print(f"⚠️  Could not load sentence-transformers: {_e}")
    raise SystemExit("Part Three cannot run without the embedding model.")

# Embed the three opinions
embedding_formalis = embedding_model.encode(opinion_formalis, convert_to_numpy=True)
embedding_purposiva = embedding_model.encode(opinion_purposiva, convert_to_numpy=True)
embedding_critica = embedding_model.encode(opinion_critica, convert_to_numpy=True)

print(f"\n✓ Opinion of Judge Formalis embedded: {embedding_formalis.shape}")
print(f"✓ Opinion of Judge Purposiva embedded: {embedding_purposiva.shape}")
print(f"✓ Opinion of Judge Critica embedded: {embedding_critica.shape}")

# Store in dict for later use
embeddings_opinions = {
    "Formalis": embedding_formalis,
    "Purposiva": embedding_purposiva,
    "Critica": embedding_critica
}
```

```python
#@title Compute Cosine Similarity Between Opinions
#@markdown Cosine similarity measures the angle between two vectors. Ranges from -1 (opposite) to +1 (identical).
#@markdown A similarity of 0.8 means the opinions are quite similar; 0.3 means they are quite different.

# Create array of embeddings
embeddings_array = np.array([
    embedding_formalis,
    embedding_purposiva,
    embedding_critica
])

# Compute pairwise cosine similarity
similarity_matrix = cosine_similarity(embeddings_array)
judge_names = ["Formalis", "Purposiva", "Critica"]

print("COSINE SIMILARITY BETWEEN JUDICIAL OPINIONS")
print("="*60)
print(f"\n{'':12} Formalis    Purposiva    Critica")
for i, name_i in enumerate(judge_names):
    print(f"{name_i:12}", end="")
    for j in range(3):
        print(f"{similarity_matrix[i][j]:8.4f}     ", end="")
    print()

print(f"\n{'─'*60}")

# Extract pairwise similarities
sim_formalis_purposiva = similarity_matrix[0][1]
sim_formalis_critica = similarity_matrix[0][2]
sim_purposiva_critica = similarity_matrix[1][2]

print(f"\nKEY FINDINGS:")
print(f"  • Formalis ↔ Purposiva: {sim_formalis_purposiva:.4f}")
print(f"  • Formalis ↔ Critica:   {sim_formalis_critica:.4f}")
print(f"  • Purposiva ↔ Critica:  {sim_purposiva_critica:.4f}")

most_similar_pair = max(
    [("Formalis", "Purposiva", sim_formalis_purposiva),
     ("Formalis", "Critica", sim_formalis_critica),
     ("Purposiva", "Critica", sim_purposiva_critica)],
    key=lambda x: x[2]
)

least_similar_pair = min(
    [("Formalis", "Purposiva", sim_formalis_purposiva),
     ("Formalis", "Critica", sim_formalis_critica),
     ("Purposiva", "Critica", sim_purposiva_critica)],
    key=lambda x: x[2]
)

print(f"\n  ✓ Most similar: {most_similar_pair[0]} & {most_similar_pair[1]} ({most_similar_pair[2]:.4f})")
print(f"  ✓ Most different: {least_similar_pair[0]} & {least_similar_pair[1]} ({least_similar_pair[2]:.4f})")
```

```python
#@title Visualize: 2D Projection of Judicial Reasoning
#@markdown We use PCA (Principal Component Analysis) to project the 384-dimensional embeddings onto 2D.
#@markdown This lets us visualize where each judge's reasoning falls in conceptual space.

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Apply PCA to reduce to 2D
pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(embeddings_array)

# Create visualization
fig, ax = plt.subplots(figsize=(10, 8))

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
judge_names_full = ['Judge Formalis\n(Textualist)', 'Judge Purposiva\n(Purposivist)', 'Judge Critica\n(TWAIL)']

# Plot each opinion
for i, (name, color) in enumerate(zip(judge_names_full, colors)):
    ax.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1], s=500, alpha=0.7, color=color, edgecolors='black', linewidth=2)
    ax.annotate(name, (embeddings_2d[i, 0], embeddings_2d[i, 1]), 
                fontsize=11, fontweight='bold', ha='center', va='center')

# Draw lines showing similarity distances
ax.plot(embeddings_2d[[0, 1], 0], embeddings_2d[[0, 1], 1], 'k--', alpha=0.3, linewidth=1)
ax.plot(embeddings_2d[[0, 2], 0], embeddings_2d[[0, 2], 1], 'k--', alpha=0.3, linewidth=1)
ax.plot(embeddings_2d[[1, 2], 0], embeddings_2d[[1, 2], 1], 'k--', alpha=0.3, linewidth=1)

ax.set_xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]:.1%} variance)', fontsize=11, fontweight='bold')
ax.set_ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]:.1%} variance)', fontsize=11, fontweight='bold')
ax.set_title('Judicial Reasoning in Vector Space\nLab 09: The AI Arbitrator', fontsize=13, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3)
ax.set_facecolor('#F8F9FA')

plt.tight_layout()
plt.savefig('judges_embedding_space.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✓ Visualization saved as 'judges_embedding_space.png'")
```

# Part Five // Reflection – The Future of Algorithmic Adjudication {#scrollTo=part_five}

## The Deeper Questions

We have now demonstrated that:

1. **Judicial reasoning is computationally modelable.** Three personas with different philosophies generated different opinions–and their differences are measurable in vector space.

2. **Legal thought has detectable patterns.** Formalist texts cluster near each other. TWAIL texts cluster together. An AI trained on these patterns can recognize and replicate them.

3. **The "Everything You Have Ever Said" framework is plausible.** If we embedded all of Judge Formalis's past opinions, they would form a distinct cluster. An AI trained on that cluster could predict her future reasoning with high accuracy.

But this technical capability raises profound legal and ethical questions:

---

## 1. Due Process and the Right to a Fair Trial

**The Question**: Can an AI arbitrate?

International human rights law guarantees the right to a fair trial:

- **ICCPR Article 14**: "Everyone shall have the right to a fair and public hearing by a competent, independent and impartial tribunal established by law."
- **ECHR Article 6**: "In the determination of his civil rights and obligations or of any criminal charge against him, everyone is entitled to a fair and public hearing within a reasonable time by an independent and impartial tribunal established by law."

**The Problem**: 

- **Independence**: Can an AI trained on past judicial opinions be truly "independent"? It is shaped by the very opinions it is supposed to transcend. It carries the biases of its training data.
- **Impartiality**: If an AI is trained on a judge's writings, it is not impartial–it is literally programmed to reason like that judge. But "reasoning like a judge" is not the same as "being open to the arguments in front of you."
- **Accountability**: Who is responsible if an AI arbitrator makes an error? The programmer? The judge whose writings trained it? The AI itself (which has no agency)?
- **Transparency**: AI predictions are often opaque (the "black box" problem). Parties have a right to understand why they lost. "The algorithm decided" is not an explanation.
- **Right to a human tribunal**: There is an argument that the right to a fair trial includes the right to be judged by a human being–someone capable of conscience, moral reasoning, and accountability.

---

## 2. Algorithmic Bias in Judicial AI

**The Question**: Whose reasoning is baked into the AI?

If we train an AI on past judicial opinions, we are training it on the collective reasoning of those judges–who are, statistically, predominantly:

- **Male** (especially in international arbitration)
- **From wealthy nations** (the Global North dominates international law)
- **Formalist and conservative** (international law has historically privileged state sovereignty and contractual certainty over expansive human rights)
- **Trained in Western legal traditions** (common law and civil law, not indigenous or customary legal systems)

**The Result**:

- An AI trained on these opinions will learn these biases. It will systematically favour:
  - Formalist over purposivist reasoning
  - State sovereignty over human rights
  - Contractual certainty over equitable justice
  - Wealthy parties over Global South claimants

- This is not a bug; it is baked into the training data.
- Unlike human judges, who can be (theoretically) persuaded to change their minds, an AI will replicate these biases perfectly and consistently.
- The biases become invisible: they hide behind the veneer of "neutral algorithm."

---

## 3. The "Everything You Have Ever Said" Problem

**The Question**: Should arbitrators worry about AI trained on their writings?

Dr Ridi's framework has a chilling implication: if your judicial reasoning is consistent (and it is–you don't reinvent your jurisprudence with every opinion), then an AI trained on your writings can predict and replicate your reasoning.

**The Risks for Arbitrators**:

1. **Predictability as a liability**: Parties can strategically select you (if you favour them) or challenge your appointment (if you disfavour them). Your reasoning is an open book.

2. **AI competition**: Why hire Judge Formalis to arbitrate when you can train an AI on her 500 opinions and get her reasoning instantly, at a fraction of the cost, without the human delays?

3. **Adversarial use**: Your opponent could train an AI on your opinions and use it to predict your reasoning, crafting their arguments to exploit your known biases.

4. **Privacy**: Your intellectual property–your unique legal reasoning–becomes data to be mined by others.

---

## 4. The EU AI Act Perspective

The **EU Artificial Intelligence Act** (in force as of 2024) classifies certain AI systems as "high-risk." Judicial AI falls into this category.

**Key Requirements for High-Risk AI**:

- **Transparency and documentation**: Providers must maintain detailed records of training data, methodology, and testing.
- **Human oversight**: High-risk systems must have meaningful human oversight and control.
- **Data quality**: Training data must be of sufficient quality to ensure the system is accurate and free from bias.
- **Bias monitoring**: Providers must conduct and document bias assessments.
- **Notification**: If a system is used to make judicial decisions, this must be disclosed to affected parties.

**The Challenge**: Meeting these requirements for judicial AI is extremely difficult. How do you ensure "bias-free" data when your training data is inherently biased? How do you provide "transparency" when the AI's decision-making is opaque? How do you ensure "human oversight" when the AI is replacing human judgment?

---

## 5. Connection to Session 10: The Lawyer's Future

Remember: in Session 8 (before this one), we explored **AI drafting**–how AI can draft contracts, pleadings, and legal documents.

Now, in Session 9, we are exploring **AI adjudication**–how AI can judge disputes.

In Session 10, the question becomes: **If AI can draft AND adjudicate, what remains for human lawyers?**

- The human lawyer as "legal advisor" may be displaced by AI that drafts better documents and predicts outcomes more accurately.
- The lawyer as "advocate" may be displaced by AI that argues more persuasively (by having learned from thousands of winning arguments).
- The lawyer as "counselor" may be displaced by AI that provides cheaper, faster legal advice.

But:

- **Human judgment**: Where do moral reasoning, ethical judgment, and conscience fit in a world of algorithmic adjudication? What is the lawyer's role in maintaining these?
- **Resistance and critique**: The lawyer may become the one who **questions the algorithm**, who asks "is this just?", who refuses to participate in unjust systems–even (especially) when the algorithm says they should.
- **Accountability**: The lawyer may become the guarantor of human accountability–the one who ensures that someone answers for injustice, not an algorithm hiding behind a "black box."

## Discussion Questions

1. **On predictability**: Is consistent judicial reasoning a virtue (predictability aids rule of law) or a vice (predictability enables gaming and bias)?

2. **On bias**: All judges are biased (consciously or unconsciously). Is an AI that transparently replicates bias preferable to a human judge who denies bias while unconsciously exercising it?

3. **On development rights**: Verdania claimed the right to development. Judge Purposiva likely favoured this. But does development automatically override all other rights? At what point does Verdania's development become Climatia's extinction?

4. **On due process**: What would due process require if AI were used in adjudication? Explainability? Human override? Consent from both parties? A right to demand a human judge?
