# Part One // Treaty as Code – Can We Encode International Law?

Let's start with a thought experiment. Smart contracts on blockchain have automated agreements–conditions encoded, execution automatic, no intermediaries. Could we do the same with treaty provisions? What happens when we try?

## The Treaty Provision

Below is a fictional but realistic treaty on emissions reductions. (It draws inspiration from the Paris Agreement, the Kyoto Protocol, and countless environmental treaties negotiated over decades.)

**Article 7 – Emissions Reduction Obligations**

1. Each Party shall ensure that its annual greenhouse gas emissions do not exceed 80% of its 2005 baseline levels by 2030.
2. Parties may offset up to 10% of their reduction obligation through verified carbon credit purchases from other Parties.
3. A Party experiencing "exceptional economic circumstances" may request a temporary exemption of up to 3 years, subject to approval by the Conference of Parties.
4. Each Party shall submit an annual report detailing its emissions levels, reduction measures, and any use of offsets or exemptions.

**The Challenge**: Encode this provision as Python code. Test compliance. Then break it.

```python
#@title Treaty Encoder: Article 7 Emissions Compliance

def check_compliance(party_name, baseline_2005, current_emissions, offsets_used=0, exemption_years=0):
    """
    Check compliance with Article 7 — Emissions Reduction Obligations.
    
    Args:
        party_name (str): Name of the party (for reporting)
        baseline_2005 (float): Baseline emissions in 2005 (in tonnes CO2e)
        current_emissions (float): Current annual emissions (in tonnes CO2e)
        offsets_used (float): Carbon credits purchased (in tonnes CO2e)
        exemption_years (int): Years of exemption granted (0-3)
    
    Returns:
        dict: compliance status, details, and reasoning
    """
    
    # Article 7(1): emissions must not exceed 80% of 2005 baseline
    emissions_cap = baseline_2005 * 0.80
    effective_emissions = current_emissions - offsets_used  # Article 7(2): offset deduction
    
    # If exemption granted, relax the constraint
    if exemption_years > 0:
        status = "EXEMPTION_GRANTED"
        reason = f"Temporary exemption approved for {exemption_years} year(s) under Article 7(3)."
        compliant = True
    else:
        # Check compliance
        if effective_emissions <= emissions_cap:
            status = "COMPLIANT"
            reason = f"Emissions {effective_emissions:.2f} ≤ cap {emissions_cap:.2f}"
            compliant = True
        else:
            status = "NON_COMPLIANT"
            reason = f"Emissions {effective_emissions:.2f} > cap {emissions_cap:.2f}"
            compliant = False
    
    return {
        "party": party_name,
        "baseline_2005": baseline_2005,
        "current_emissions": current_emissions,
        "emissions_cap": emissions_cap,
        "offsets_used": offsets_used,
        "effective_emissions": effective_emissions,
        "exemption_years": exemption_years,
        "status": status,
        "reason": reason,
        "compliant": compliant
    }

print("✓ Treaty encoder loaded. Ready to test compliance scenarios.")
```

```python
#@title Test Case 1: Straightforward Compliance
#@markdown Testing three hypothetical parties with clear outcomes.

test_cases = [
    {"name": "Party A", "baseline": 1000.0, "current": 750.0, "offsets": 0, "exemption": 0},
    {"name": "Party B", "baseline": 1000.0, "current": 850.0, "offsets": 80.0, "exemption": 0},
    {"name": "Party C", "baseline": 1000.0, "current": 900.0, "offsets": 0, "exemption": 0},
]

print("\n" + "="*80)
print("COMPLIANCE TESTING: STRAIGHTFORWARD CASES")
print("="*80)

for test in test_cases:
    result = check_compliance(
        test["name"],
        test["baseline"],
        test["current"],
        test["offsets"],
        test["exemption"]
    )
    
    print(f"\n{test['name']}:")
    print(f"  Baseline (2005):        {result['baseline_2005']:.2f} tonnes CO2e")
    print(f"  Current emissions:      {result['current_emissions']:.2f} tonnes CO2e")
    print(f"  80% cap (Article 7.1):  {result['emissions_cap']:.2f} tonnes CO2e")
    print(f"  Offsets used:           {result['offsets_used']:.2f} tonnes CO2e")
    print(f"  Effective emissions:    {result['effective_emissions']:.2f} tonnes CO2e")
    print(f"  ➜ STATUS: {result['status']}")
    print(f"     {result['reason']}")

print("\n" + "="*80)
```

# Part Two // Where Code Breaks – The Vagueness Feature

The treaty reads cleanly. The code compiles. The tests pass. But *real-world treaty implementation* is messier. What happens when we test edge cases?

## Edge Case 1: Temporal Ambiguity – What is "Annual"?

```python
#@title Edge Case: Temporal Ambiguity

print("\n" + "="*80)
print("EDGE CASE 1: WHAT DOES 'ANNUAL' MEAN?")
print("="*80)

print("""
The treaty says: "Each Party shall ensure that its annual greenhouse gas 
emissions do not exceed 80% of its 2005 baseline levels by 2030."

But what is 'annual'? The code assumes calendar year (Jan–Dec). But:
  • A fiscal year country (e.g., India: Apr–Mar) would measure differently.
  • Some systems use rolling 12-month averages.
  • Some use financial reporting years.

Consequence: The SAME emissions data could be compliant or non-compliant 
depending on which calendar you use.

Legal reality: Treaties often leave this ambiguous. The Conference of Parties 
will issue guidance—but code demands precision NOW.
""")

print("\nExample:")
print("  Country reports 800 tonnes in calendar year 2030.")
print("  Cap is 800 tonnes (80% of 1000).")
print("  Calendar year: EXACTLY AT CAP (compliant? rounding error?)")
print("  Fiscal year shifted by 1 month: different measurement period.")
print("\n→ Code returned COMPLIANT. But should it? The treaty doesn't say.")
```

## Edge Case 2: The Vagueness Feature – "Exceptional Economic Circumstances"

```python
#@title Edge Case: Vagueness as a Feature

print("\n" + "="*80)
print("EDGE CASE 2: 'EXCEPTIONAL ECONOMIC CIRCUMSTANCES'")
print("="*80)

print("""
Article 7(3) says: "A Party experiencing 'exceptional economic circumstances' 
may request a temporary exemption of up to 3 years."

But what counts as 'exceptional'?
  • A 2% GDP decline? 5%? 15%?
  • Does unemployment matter? Inflation?
  • What about currency crises?
  • War? Natural disaster? Pandemic?
  • Does 'exceptional' mean rare, or does a country get to claim it every decade?

The code above treats exemption as a boolean flag (exemption_years > 0).
But the treaty uses VAGUE LANGUAGE deliberately.

Why? Because the drafters WANT diplomatic flexibility. The Conference of Parties 
will judge each case on its merits. Good faith negotiation. No formula can capture 
all scenarios.

Consequence: The vagueness is NOT a drafting failure—it's a FEATURE.
It allows the treaty to adapt to unforeseen circumstances. But code demands YES/NO.
""")

print("\n⚠️  THOUGHT EXPERIMENT:")
print("    If we hard-code a definition (e.g., 'GDP decline > 5%'),")
print("    we solve the problem for normal cases but create absurd edge cases:")
print("      • A country just barely under 5% decline: no exemption.")
print("      • A country just barely over 5%: exemption granted.")
print("      • A tiny island nation with 5.01% decline gets same exemption as Germany.")
print("\n    This mechanical logic would likely be rejected by diplomats as unfair.")
```

## Edge Case 3: Floating-Point Precision – The Boundary Problem

```python
#@title Edge Case: Floating-Point Precision at the Boundary

print("\n" + "="*80)
print("EDGE CASE 3: FLOATING-POINT PRECISION IN LAW")
print("="*80)

# The classic floating-point problem
print("\nClassic problem: 0.1 + 0.2 in binary floating-point:")
print(f"  0.1 + 0.2 = {0.1 + 0.2}")
print(f"  0.1 + 0.2 == 0.3? {0.1 + 0.2 == 0.3}")
print(f"  ✗ Result: False (due to binary representation)")

print("\n" + "-"*80)
print("\nNow: apply this to treaty compliance.")

# Scenario
baseline = 1000.0
cap = baseline * 0.80  # Should be 800.0

print(f"\n  Baseline: {baseline}")
print(f"  80% of baseline: {cap}")
print(f"  Stored cap value: {repr(cap)}")

emissions = 800.0
print(f"  Reported emissions: {emissions}")
print(f"  emissions <= cap? {emissions <= cap}")

print("\n✓ That worked. But what if:")

# A more insidious case
emissions2 = 799.9999999999999
print(f"\n  Reported emissions: {emissions2}")
print(f"  emissions <= cap? {emissions2 <= cap}")
print(f"  Difference: {cap - emissions2}")

print("\n⚠️  IMPLICATION:")
print("    A country 0.0000000001 tonnes BELOW the cap is compliant.")
print("    But a country 0.0000001 ABOVE might fail due to rounding errors")
print("    in how emissions were measured or reported.")
print("\n    Treaty language says 'does not exceed 80%.'")
print("    But at what precision? 1 tonne? 0.1 tonne? 1 kg?")
print("    The treaty leaves this ambiguous. Code forces a decision.")
```

## Edge Case 4: Offset Verification – What is "Verified"?

```python
#@title Edge Case: Offset Verification Status

print("\n" + "="*80)
print("EDGE CASE 4: OFFSET VERIFICATION")
print("="*80)

print("""
Article 7(2): "Parties may offset up to 10% of their reduction obligation through 
verified carbon credit purchases from other Parties."

Key word: 'verified.' Our code assumes offsets are binary: used or not used.
But the real timeline:
  • T=0: Country A buys credits from Country B.
  • T=1-6 months: Verification process (audits, inspections, disputes).
  • T=6+ months: Credits verified. Offset counts toward compliance.
  • T=unknown: What if verification fails halfway through?

Our code (line: effective_emissions = current_emissions - offsets_used)
assumes all offsets are "verified" at the moment of reporting.

But what if:
  • A credit verification is challenged?
  • The selling country later loses accreditation?
  • The carbon project is found to be fraudulent?
  • Verification is pending at reporting deadline?

Treaty answer: "This is a matter for the Conference of Parties to decide." (Vague!)
Code answer: Boolean flag at report time. (Precise!)
They don't align.
""")

# Enhanced compliance check with offset verification
def check_compliance_with_offset_verification(
    party_name, baseline_2005, current_emissions, 
    offsets_pending=0, offsets_verified=0, exemption_years=0
):
    """
    Enhanced version distinguishing between verified and pending offsets.
    """
    emissions_cap = baseline_2005 * 0.80
    
    # Only verified offsets count toward compliance
    effective_emissions = current_emissions - offsets_verified
    
    if exemption_years > 0:
        status = "EXEMPTION_GRANTED"
        reason = f"Exemption for {exemption_years} year(s); pending offsets: {offsets_pending}"
        compliant = True
    else:
        if effective_emissions <= emissions_cap:
            status = "COMPLIANT"
            reason = f"Verified offsets: {offsets_verified}; pending: {offsets_pending}"
            compliant = True
        else:
            status = "NON_COMPLIANT"
            reason = f"Even with verified offsets, still exceeds cap."
            compliant = False
    
    return {
        "party": party_name,
        "current_emissions": current_emissions,
        "offsets_verified": offsets_verified,
        "offsets_pending": offsets_pending,
        "effective_emissions": effective_emissions,
        "status": status,
        "reason": reason
    }

print("\nExample: Country D claims 850 tonnes with 80 tonnes of offsets.")
result = check_compliance_with_offset_verification("Country D", 1000, 850, 50, 30, 0)
print(f"  50 tonnes verified, 50 tonnes pending.")
print(f"  → Status: {result['status']} ({result['reason']})")
print(f"\nOur code made a choice: only verified offsets count. ✓ Reasonable.")
print(f"But the treaty is SILENT on this detail.")
```

## Key Insight: Vagueness as Diplomatic Flexibility

**The Big Idea**: The vagueness in Article 7 is *not a drafting failure*. It's a **feature**.

Why do treaties use vague language?

1. **Unforeseen Circumstances**: The drafters cannot predict all future scenarios. Vagueness allows the treaty to adapt.
2. **Good-Faith Negotiation**: Rather than lock in rigid rules, treaties often create frameworks for parties to negotiate, case-by-case, what counts as compliance in novel situations.
3. **Distributional Fairness**: A fixed rule ("GDP decline > 5%") helps some countries, harms others. Vagueness allows equity considerations.
4. **Institutional Learning**: Over time, the Conference of Parties develops *practice* and *interpretation*, sharpening the rule through experience.

**Code demands precision.** Law often requires *interpreted flexibility*.

This tension–between computational logic and legal interpretation–is at the heart of computational law.

# Part Three // Computational Analysis of Custom – NLP Meets the ICJ

Now we shift perspective. Instead of asking "can we encode law as code," we ask: "can we use computation to understand how law is *made*?"

Customary international law (CIL) is made through state practice and opinio juris. But how do courts *identify* CIL? The International Court of Justice (ICJ) is the authoritative body. A computational analysis of 100+ ICJ judgments reveals that the Court cites **its own precedent (54.9%)** far more often than it identifies fresh state practice. This raises a profound question: *Is the ICJ making law, or finding it?*

## NLP Pipeline: Detecting CIL Propositions

```python
#@title CIL Extraction Pipeline

def extract_cil_indicators(text):
    """
    Step 1: Detect keywords/phrases indicating CIL analysis.
    """
    indicators = {
        "customary": r"customary\s+international\s+law|customary\s+law",
        "state_practice": r"state\s+practice|general\s+practice",
        "opinio_juris": r"opinio\s+juris|legal\s+conviction|accepted\s+as\s+law",
        "general_principle": r"general\s+principle|fundamental\s+rule",
        "judicial_precedent": r"jurisprudence|precedent|court|judgment",
        "uniform_practice": r"uniform\s+practice|constant\s+usage|consistent"
    }
    
    text_lower = text.lower()
    found = {}
    for key, pattern in indicators.items():
        matches = re.findall(pattern, text_lower)
        if matches:
            found[key] = len(matches)
    
    return found

def segment_sentences(text):
    """
    Step 2: Sentence segmentation using spaCy.
    """
    doc = nlp(text)
    return [sent.text.strip() for sent in doc.sents]

def analyze_icj_excerpt(case_name, excerpt):
    """
    Full pipeline: analyze an ICJ excerpt for CIL content.
    """
    # Segment sentences
    sentences = segment_sentences(excerpt)
    
    # Extract indicators
    indicators = extract_cil_indicators(excerpt)
    
    # Analyze sentences for CIL signals
    cil_sentences = []
    for sent in sentences:
        sent_indicators = extract_cil_indicators(sent)
        if sent_indicators:
            cil_sentences.append({"text": sent, "signals": sent_indicators})
    
    # Source attribution (rule-based)
    source = "unknown"
    if "jurisprudence" in indicators or "precedent" in indicators or "judgment" in indicators:
        source = "judicial_precedent"
    if "state_practice" in indicators or "uniform_practice" in indicators:
        source = "state_practice"
    if "opinio_juris" in indicators:
        source = "opinio_juris"
    if "general_principle" in indicators:
        source = "general_principle"
    
    return {
        "case": case_name,
        "num_sentences": len(sentences),
        "num_cil_sentences": len(cil_sentences),
        "indicators": indicators,
        "cil_sentences": cil_sentences,
        "primary_source": source
    }

print("✓ NLP pipeline defined.")
```

```python
#@title Run Pipeline on Corpus

results = []
for item in icj_corpus:
    result = analyze_icj_excerpt(item["case"], item["excerpt"])
    result["ground_truth_cil"] = item["has_cil"]
    results.append(result)

print("\n" + "="*80)
print("CIL DETECTION RESULTS")
print("="*80)

for i, res in enumerate(results, 1):
    print(f"\n{i}. {res['case']}")
    print(f"   Sentences: {res['num_sentences']} total, {res['num_cil_sentences']} with CIL signals")
    print(f"   Indicators: {res['indicators']}")
    print(f"   Primary source: {res['primary_source']}")
    if res['ground_truth_cil']:
        print(f"   ✓ Ground truth: Has CIL content")
    else:
        print(f"   ✗ Ground truth: No CIL content")
```

## Source Attribution: What Does the ICJ Cite?

```python
#@title Source Attribution Analysis

import matplotlib.pyplot as plt
from collections import Counter

# Count source types
sources = [r["primary_source"] for r in results]
source_counts = Counter(sources)

print("\n" + "="*80)
print("SOURCE ATTRIBUTION: WHERE DOES THE ICJ CITE?")
print("="*80)

print(f"\nBreakdown (n={len(results)} excerpts):")
for source, count in source_counts.most_common():
    pct = 100 * count / len(results)
    print(f"  {source:25s}: {count:2d} ({pct:5.1f}%)")

print("\n" + "-"*80)
print("\nKey Finding (Ridi 2024):")
print("  The ICJ cites its own jurisprudence (judicial precedent) far more")
print("  often than it identifies fresh state practice. This suggests that")
print("  CIL is constituted not by finding practice, but by the Court's")
print("  *interpretation and synthesis* of prior case law.")
print("\n  This challenges traditional CIL doctrine (ICJ Statute Article 38),")
print("  which treats state practice + opinio juris as the sources, with the")
print("  Court merely applying them.")

# Visualization
fig, ax = plt.subplots(figsize=(8, 6))
source_labels = [s.replace("_", "\n") for s in source_counts.keys()]
source_values = source_counts.values()
colors = ["#FF6B6B", "#4ECDC4", "#45B7D1", "#FFA07A", "#98D8C8"]
ax.pie(source_values, labels=source_labels, autopct="%1.1f%%", colors=colors[:len(source_labels)])
ax.set_title("ICJ Source Attribution in CIL Analysis\n(Simplified corpus, n=8)")
plt.tight_layout()
plt.show()

print("\n(Note: This small corpus shows patterns; the full analysis includes 100+ judgments)")
```

# Part Four // The Headnote Harvester Preview – Scaling Legal Analysis

The CIL pipeline above is manual and small-scale. Real computational legal research requires *automated* extraction at scale. Here's a preview of one approach: the **Headnote Harvester**, which uses OCR + NLP to extract legal propositions from thousands of cases.

## Named Entity Recognition in Legal Text

```python
#@title Entity Extraction: Building a Legal Knowledge Graph

def extract_legal_entities(text):
    """
    Use spaCy NER to extract entities from legal text.
    """
    doc = nlp(text)
    entities = {"PERSON": [], "GPE": [], "ORG": [], "DATE": []}
    
    for ent in doc.ents:
        if ent.label_ in entities:
            entities[ent.label_].append(ent.text)
    
    return entities

print("\n" + "="*80)
print("ENTITY EXTRACTION: KNOWLEDGE GRAPH CONSTRUCTION")
print("="*80)

# Example: extract from a judgment
sample_text = """In the case of Nicaragua v. United States of America, 
the International Court of Justice, sitting at The Hague on 27 June 1986, 
held that the use of force violated customary international law."""

print(f"\nSample text: {sample_text}\n")

entities = extract_legal_entities(sample_text)
for ent_type, ent_list in entities.items():
    if ent_list:
        print(f"  {ent_type:10s}: {', '.join(set(ent_list))}")

print("\n" + "-"*80)
print("\nHow this scales to a knowledge graph:")
print("  1. Extract entities from 1,000s of cases.")
print("  2. Link cases via shared entities (countries, courts, dates).")
print("  3. Infer relationships: Case A cites Case B → edge in graph.")
print("  4. Analyze graph structure: Which cases are most cited? Which countries?")
print("\nThis is the foundation of the Headnote Harvester system.")
print("(Full implementation in a separate notebook.)")
```

# Part Five // Reflection – Should Law Be Computable?

We've now seen both sides of the computational law question:

1. **Treaty as Code**: Encoding Article 7 shows that law *can* be partially expressed as code–but vagueness, edge cases, and ambiguity in human language resist formalization.
2. **Custom as Pattern**: NLP reveals patterns in how courts *make* law–the ICJ constitutes custom largely through self-citation, not fresh state practice.

These observations raise deeper questions. Let's reflect on them.

## Discussion: The Incomplete Contracts Framework

```python
#@title The Economics of Vagueness

print("""
╔══════════════════════════════════════════════════════════════════════════════╗
║                  INCOMPLETE CONTRACTS FRAMEWORK                              ║
║                                                                              ║
║  Economic theory (Grossman-Hart, Hart-Moore) explains why contracts are      ║
║  intentionally vague or incomplete:                                          ║
║                                                                              ║
║  • Complete Specification is Costly: Drafting every contingency costs time   ║
║    and money. At some point, more precision is not worth the cost.           ║
║                                                                              ║
║  • Unforeseen Circumstances: Some scenarios cannot be foreseen. Vagueness    ║
║    allows the contract to adapt.                                            ║
║                                                                              ║
║  • Adaptation via Renegotiation: When unexpected events occur, parties can  ║
║    renegotiate in good faith. Rigid rules prevent mutual gain from           ║
║    adaptation.                                                              ║
║                                                                              ║
║  • Power Allocation: Vague terms delegate power to courts or arbitrators     ║
║    (or to the Conference of Parties). This can be efficient if those         ║
║    decision-makers are trustworthy.                                         ║
║                                                                              ║
║  APPLIED TO TREATIES:                                                        ║
║  • Treaty negotiation is expensive. Specifying every rule in advance         ║
║    ("3 years exemption if GDP falls >5%") creates rigidity.                  ║
║  • The drafters KNOW they don't know the future.                             ║
║  • So they write vague terms ("exceptional circumstances") and delegate       ║
║    interpretation to the Conference of Parties.                              ║
║  • This is EFFICIENT, not sloppy.                                            ║
║                                                                              ║
║  BUT: This efficiency comes at a cost.                                      ║
║  • If the Conference of Parties is corrupted, vagueness becomes a tool for   ║
║    exploitation.                                                            ║
║  • Rule of law requires predictability. Vagueness can undermine that.        ║
║                                                                              ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")
```

## Discussion: Computational Law vs. Rule of Law

```python
#@title Computational Law and Governance

print("""
╔══════════════════════════════════════════════════════════════════════════════╗
║         IF RULES ARE CODE, WHO CONTROLS THE CODE?                           ║
║                                                                              ║
║  The move from "treaties as interpreted by diplomats" to "rules as          ║
║  executable code" shifts power:                                             ║
║                                                                              ║
║  TRADITIONAL:                                                               ║
║    Treaty text → Interpreted by diplomats at COP → Flexible outcomes        ║
║                                                                              ║
║  COMPUTATIONAL:                                                              ║
║    Code → Executed by algorithm → Deterministic outcomes                    ║
║                                                                              ║
║  CONTROL QUESTION: Who writes the code? Who audits it? Who can update it?  ║
║                                                                              ║
║  EXAMPLES:                                                                  ║
║                                                                              ║
║  1. SMART TREATIES ON BLOCKCHAIN:                                           ║
║     • Emissions trading: if Party A sells credits to Party B, payment       ║
║       and delivery execute automatically, no human intermediaries.           ║
║     • Benefit: No corruption, instant settlement.                          ║
║     • Risk: Code bug = financial loss. Who's liable?                        ║
║     • Risk: Code immutability = cannot adapt to new circumstances.          ║
║                                                                              ║
║  2. AI ADJUDICATION:                                                        ║
║     • Algorithm classifies compliance: compliant or non-compliant.           ║
║     • Benefit: Speed, consistency.                                          ║
║     • Risk: Algorithm bias. If training data is skewed, outcomes are        ║
║       systematically unfair to some nations.                                ║
║     • Rule of law requires JUSTIFICATION. Code provides opaque decisions.    ║
║                                                                              ║
║  3. ENVIRONMENTAL MONITORING:                                               ║
║     • Satellite data + ML: detect illegal fishing, unauthorized logging.    ║
║     • Benefit: Evidence-based enforcement at scale.                         ║
║     • Risk: False positives harm innocent communities. Who appeals?          ║
║                                                                              ║
║  THE DEEPER ISSUE:                                                          ║
║  • Rule of law requires: legality (rules known in advance), equality,       ║
║    participation, accountability.                                          ║
║  • Computational law offers efficiency but threatens accountability:         ║
║    How do you appeal a decision made by an algorithm?                      ║
║    How do developing nations without tech infrastructure participate?        ║
║                                                                              ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")
```

## Discussion: Desirable Domains for Computational Law

```python
#@title Which Areas of Law Benefit Most from Encoding?

print("""
╔══════════════════════════════════════════════════════════════════════════════╗
║           COMPARATIVE ADVANTAGE: WHEN IS COMPUTATION VALUABLE?               ║
║                                                                              ║
║  STRONG CANDIDATES FOR ENCODING:                                            ║
║                                                                              ║
║  1. ENVIRONMENTAL TREATIES (e.g., emissions, fisheries)                      ║
║     • Why: Monitoring is objective (emissions data from satellites).         ║
║     • Benefit: Real-time compliance tracking. Early warning for violations.  ║
║     • Caution: Precision doesn't equal fairness. Still need COP approval.   ║
║                                                                              ║
║  2. TECHNICAL STANDARDS (e.g., labor conditions, product safety)             ║
║     • Why: Rules are already detailed ("pH < 7.0", "tensile strength > 50 MPa").   ║
║     • Benefit: Automated inspection + certification.                        ║
║     • Caution: Who sets the standard? Whose interest does it serve?         ║
║                                                                              ║
║  3. DISPUTE RESOLUTION (e.g., WTO appeals, arbitration)                      ║
║     • Why: Computational analysis of precedent can predict outcomes.         ║
║     • Benefit: Faster, cheaper dispute resolution.                          ║
║     • Caution: Reduces opportunity for novel legal reasoning + equity.      ║
║                                                                              ║
║  WEAK CANDIDATES (or dangerous):                                            ║
║                                                                              ║
║  1. HUMAN RIGHTS LAW                                                        ║
║     • Why: Violations are contextual, inherently subjective.                 ║
║     • Problem: "Torture" cannot be defined as a mathematical formula.        ║
║     • Problem: Algorithmic classification risks institutionalizing bias.     ║
║     • Example: Refugee status decisions based on ML = systematic exclusion?  ║
║                                                                              ║
║  2. DIPLOMATIC NEGOTIATIONS                                                 ║
║     • Why: Core function is human judgment + good faith.                    ║
║     • Problem: Vagueness is a FEATURE (as we've seen).                      ║
║     • Problem: If rules are rigid code, fewer opportunities for creativity. ║
║                                                                              ║
║  3. TREATY DRAFTING                                                         ║
║     • Why: Requires balancing incommensurable values (sovereignty vs. 
║       environment, development vs. conservation).                           ║
║     • Problem: No algorithm can balance values. This is inherently political.║
║                                                                              ║
║  HYBRID APPROACH (LIKELY THE FUTURE):                                       ║
║  • Use computation for evidence gathering (satellite data, statistics).      ║
║  • Use computation for proposal generation ("here are N treaty amendments    ║
║    that maximize Pareto-efficiency").                                       ║
║  • Keep humans in the loop for contested decisions, especially those        ║
║    affecting rights or fairness.                                            ║
║                                                                              ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")
```

## The Empirical Turn: What Can Computation Reveal?

```python
#@title The Empirical Shift in International Law Scholarship

print("""
╔══════════════════════════════════════════════════════════════════════════════╗
║      FROM ARMCHAIR THEORY TO COMPUTATIONAL EMPIRICISM                        ║
║                                                                              ║
║  Traditional International Law Scholarship:                                 ║
║    • Lawyer reads treaties + cases → forms an intuitive view of the law.   ║
║    • Analysis is hermeneutic (interpretive) and particular.                  ║
║    • Claims are supported by examples, not statistics.                      ║
║    • Questions: What does this case mean? How should it be applied?         ║
║                                                                              ║
║  Computational International Law Scholarship:                               ║
║    • Researcher extracts patterns from 100s-1000s of cases via NLP.         ║
║    • Analysis is quantitative and general.                                  ║
║    • Claims are supported by statistical evidence.                          ║
║    • Questions: What patterns emerge at scale? What contradicts doctrine?    ║
║                                                                              ║
║  EXAMPLE: Ridi (2024) on Customary International Law                        ║
║    • Traditional view: ICJ identifies CIL by examining state practice +      ║
║      opinio juris (ICJ Statute Article 38).                                ║
║    • Empirical finding: ICJ cites its own prior judgments 54.9% of the      ║
║      time, far more than fresh evidence of state practice.                 ║
║    • Implication: CIL is not being "found" but "made" by the Court.        ║
║    • This contradicts traditional doctrine but is hard to see from          ║
║      reading a few cases in isolation.                                     ║
║                                                                              ║
║  WHAT COMPUTATION REVEALS (that close reading cannot):                      ║
║    • Aggregate patterns across large bodies of law.                          ║
║    • Contradictions between doctrine and practice.                           ║
║    • Temporal trends (is the law becoming more lenient? more strict?).       ║
║    • Disparity: Are some nations treated differently?                        ║
║    • Precedential importance: Which cases are most cited? Why?               ║
║                                                                              ║
║  LIMITATIONS:                                                               ║
║    • Computation finds correlations, not causation.                          ║
║    • Quantitative analysis misses nuance and context.                        ║
║    • Risk of "quantifying the unquantifiable" (values, fairness, justice).   ║
║                                                                              ║
║  BEST PRACTICE: Combine computational analysis with traditional scholarship  ║
║    Computation as a tool for hypothesis generation, not final authority.     ║
║                                                                              ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")
```
