# Part One: The Treaty Landscape {#scrollTo=part-one}

#@markdown ### The AI Governance Landscape Is Fragmented

As of 2026, there is **no single comprehensive international treaty on AI**. Instead, the world has inherited a patchwork of instruments, each reflecting different values, jurisdictions, and approaches:

- The **EU AI Act** (2024): A regulatory gold standard, but regional and legally binding only within the EU.
- The **Council of Europe Framework Convention on AI** (2024): A multilateral treaty, but consensus-based and weaker in enforcement.
- The **OECD AI Principles** (2019): Influential, but non-binding. Reflect the values of wealthy democracies.
- The **Hiroshima Process** (2023–onwards): A temporary political commitment by G7 states. No binding force.
- The **Global Digital Compact** (2024): A UN initiative, but vague and declaratory.
- Scattered **bilateral agreements** (US–UK on AI safety, for example).
- **National legislation** in countries like Singapore, Brazil, Canada—each with different standards.

This is regime complexity in action. Recall **Rainer Maas's framework** from Session 5: the world has chosen (or stumbled into) **managed pluralism**—multiple overlapping regimes, some binding, some soft, some purely directional. This creates both opportunities and risks:

**Opportunities:**
- Regulatory experimentation (what Maas calls "experimentalist governance").
- Different communities can adopt different standards.
- Flexibility as knowledge about AI risks evolves.

**Risks:**
- Regulatory arbitrage (companies shop for the friendliest regime).
- Incoherence (what is "AI transparency" under the EU Act vs. the CoE Convention?).
- Exclusion (powerful states and corporations set the agenda; weaker states and marginalized communities follow).

#@markdown ### What's Missing? Where Are the Gaps?

Despite this flurry of activity, significant gaps remain:

1. **No global binding treaty on AI** comparable to treaties on nuclear non-proliferation, climate, or biodiversity.

2. **Autonomous weapons systems** remain largely unregulated at the international level. The Humanitarian Law and AI project at the UN is ongoing, but no binding protocol exists.

3. **Cross-border data flows and digital sovereignty** are fragmented. The EU's GDPR and data transfer mechanisms coexist with Balkanized national rules (India's data localization, China's internet sovereignty doctrine). No unified international standard.

4. **Environmental impact of AI** is barely addressed. Training large models is energy-intensive. Rare earth mining for semiconductors is ecologically destructive. No treaty provisions on this.

5. **AI and human rights** have soft-law commitments (UNESCO, OHCHR) but no binding international enforcement mechanism.

6. **Liability and accountability** across borders remain unclear. If an AI system trained in one country causes harm in another, which law applies? Who is liable?

7. **Meaningful human control in military contexts**—widely advocated but never formalized in a binding treaty.

8. **The Global South's voice** is underrepresented. Most instruments were drafted by wealthy democracies and reflected their interests and technological capacities.

Your task today is to help draft a provision that addresses one of these gaps. You will use AI as a partner in that drafting. And in doing so, you will enact—in compressed form—the real negotiation, critique, and revision that would occur if the UN General Assembly actually convened a preparatory committee for a Framework Convention on AI.

# Part Three: AI-Assisted Drafting — Round 1 {#scrollTo=part-three}

#@markdown ### The First Draft

Now you will use Gemini to generate a first draft of your article. The AI has been given a system prompt that teaches it to write like a treaty drafter:

- Clear numbered paragraphs
- Formal language: "Each Party shall...", "The States Parties shall..."
- Definitions sections (terms in *italics*)
- Recognizing that treaties need flexibility ("as appropriate", "to the extent") but not so much that enforcement becomes impossible

The AI will generate a draft based on your chosen topic. Your job is **not** to accept it as final, but to read it critically. Does it make sense? Does it answer the core question? What's missing? What's overreaching?

That analysis begins in Part Four, where you will hear from four different critics.

```python
#@title ## Generate First Draft

#@param {type:"string", help: "Any additional context or priorities for your article? (Optional)"} student_priorities

system_prompt = """You are an expert international law drafter with 20 years of experience in multilateral treaty negotiations. Your task is to draft a provision for a UN Framework Convention on Artificial Intelligence. 

Your style should follow these conventions:
1. Number paragraphs as "1.", "2.", "3.", etc.
2. Use formal diplomatic language: "Each Party shall", "The States Parties shall", "For the purposes of this Convention"
3. Include a definitions section at the start, defining key terms in italics.
4. Use nested numbering for complex obligations: "1. Each Party shall... (a) ..., (b) ..., (c) ..."
5. Build in flexibility where appropriate: "to the extent possible", "as appropriate", "in accordance with their legal systems"
6. Avoid jargon, but use precise legal terms.
7. The provision should be 4-6 paragraphs long.
8. Consider enforceability: how would a compliance committee monitor this?
9. Be mindful of capacity constraints: developing states may not have the resources to implement complex obligations.
10. Anticipate multiple interpretations: use clear, specific language to minimize ambiguity.
"""

user_prompt = f"""Draft Article X of the UN Framework Convention on Artificial Intelligence on the topic: {article_topic}.

Student priorities: {student_priorities if student_priorities else 'No specific priorities provided.'}

Please produce a complete article (4-6 paragraphs) that could realistically appear in a multilateral treaty. Include definitions, obligations, and flexibility clauses."""

print(f"Generating first draft on: {article_topic}\n")
print("="*80)

try:
    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=user_prompt,
        config=types.GenerateContentConfig(
            max_output_tokens=1500,
            system_instruction=system_prompt,
            temperature=0.7,
            thinking_config=types.ThinkingConfig(thinking_level="MINIMAL")
        )
    )
    
    first_draft = response.text
    
    print("\n" + "="*80)
    print("FIRST DRAFT")
    print("="*80 + "\n")
    print(first_draft)
    print("\n" + "="*80)
    
except Exception as e:
    print(f"Error generating draft: {e}")
    first_draft = None
```

# Part Four: Critique and Red-Teaming {#scrollTo=part-four}

#@markdown ### The Four Personas

International treaties are negotiated by diverse stakeholders. Each brings different priorities, expertise, and suspicions. Here, you will hear from four archetypical voices in any treaty negotiation:

1. **The TWAIL Scholar**: Critical legal theorist. Looks for hidden power asymmetries, colonial assumptions, impacts on Global South.
2. **The Industry Representative**: Technology sector perspective. Focused on feasibility, cost, innovation impact.
3. **The Human Rights Advocate**: Civil society voice. Demands robust protections, enforcement mechanisms, remedies.
4. **The International Law Formalist**: Jurisprudential perspective. Identifies ambiguities, inconsistencies, gaps in legal reasoning.

Each will read your draft and offer critique. Their disagreements will reveal the real tensions in treaty negotiations.

Note: This is a *simulation*, but not a false one. These are genuine perspectives that appear in actual negotiations. The only difference here is that they are articulated by an AI playing roles, rather than by humans in a Geneva conference room.

```python
#@title ## Critique 1: The TWAIL Scholar

if first_draft is None:
    print("Please generate a first draft before proceeding.")
else:
    twail_system = """You are a leading scholar of Third World Approaches to International Law (TWAIL). Your work emphasizes:
1. Historical power asymmetries embedded in international law
2. The agency and rights of Global South nations
3. Critiques of Western-centric frameworks in governance
4. The distributional impacts of rules: who benefits, who is burdened?
5. The role of colonialism in shaping contemporary legal structures

You are reading a draft treaty article on AI governance. Your role is to identify:
- Assumptions that privilege wealthy, technologically advanced states
- Obligations that may be impossible for developing nations to implement
- Who controls the knowledge production (research, data, algorithms)?
- Whose interests are served by this framing of the problem?
- What voices are missing from the negotiation?

Your critique should be substantive, specific, and constructive."""
    
    twail_prompt = f"""Read this draft article on {article_topic} for a UN Framework Convention on AI. Offer a TWAIL critique.

DRAFT ARTICLE:
{first_draft}

Provide a 250-300 word critique that identifies:
1. Implicit assumptions favoring wealthy states or tech companies
2. Implementation burdens for developing nations
3. Missing perspectives or voices
4. Power imbalances reflected in the text
5. Recommendations for revision"""
    
    try:
        response = client.models.generate_content(
            model="gemini-3-flash-preview",
            contents=twail_prompt,
            config=types.GenerateContentConfig(
                max_output_tokens=400,
                system_instruction=twail_system,
                temperature=0.8,
                thinking_config=types.ThinkingConfig(thinking_level="MINIMAL")
            )
        )
        
        critique_twail = response.text
        
        print("\n" + "="*80)
        print("CRITIQUE 1: TWAIL SCHOLAR")
        print(""""The perspective of Global South advocates, post-colonial theory, and justice""")
        print("="*80 + "\n")
        print(critique_twail)
        print("\n" + "="*80)
        
    except Exception as e:
        print(f"Error generating TWAIL critique: {e}")
```

# Part Six: Computational Analysis of Your Draft {#scrollTo=part-six}

#@markdown ### Beyond the Qualitative: NLP and Embeddings

You have now drafted, critiqued, and revised a provision. But there's another lens: the computational one.

In Sessions 2, 7, and 9, you learned about:

- **Tokenization and embeddings** (Session 2): How to represent text as numerical vectors
- **Vagueness in legal text** (Session 7): How formalisation and computation reveal ambiguity
- **Bias in AI systems** (Session 9): How embeddings encode assumptions

Now you will apply these techniques to your own draft. You will:

1. **Extract key entities and concepts** from your draft using spaCy (named entity recognition, noun phrases)
2. **Embed your draft** alongside existing treaty texts
3. **Measure similarity** to existing instruments: EU AI Act, Council of Europe Convention, OECD Principles
4. **Visualize** where your draft sits in "treaty space"

The question: **What is the genealogy of your draft?** Is it closer to the EU's approach? The OECD's? A hybrid?

```python
#@title ## Step 1: Install and Load NLP Models

import subprocess
import sys

print("Installing spacy model...")
subprocess.check_call([sys.executable, "-m", "spacy", "download", "en_core_web_sm"])

import spacy
nlp = spacy.load("en_core_web_sm")

print("✓ spaCy model loaded.\n")
```

```python
#@title ## Step 2: Extract Key Entities and Concepts

if revised_draft is None:
    print("Please generate a revised draft before proceeding.")
else:
    # Use the first paragraph or first 500 tokens for analysis
    draft_text = revised_draft.split('EXPLANATION')[0] if 'EXPLANATION' in revised_draft else revised_draft
    
    doc = nlp(draft_text[:2000])  # Process first 2000 chars to avoid timeouts
    
    # Extract entities
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    
    # Extract noun chunks (key concepts)
    noun_phrases = [chunk.text for chunk in doc.noun_chunks]
    
    print("="*80)
    print("KEY ENTITIES & CONCEPTS IN YOUR DRAFT")
    print("="*80 + "\n")
    
    print("Named Entities (organizations, dates, etc.):")
    for ent, label in entities[:10]:  # First 10
        print(f"  - {ent} ({label})")
    
    print("\nKey Noun Phrases (concepts):")
    # Deduplicate and show top phrases
    from collections import Counter
    phrase_counts = Counter(noun_phrases)
    for phrase, count in phrase_counts.most_common(10):
        print(f"  - {phrase} (appears {count}x)")
    
    print("\n" + "="*80)
    print("INTERPRETATION: These are the building blocks of your treaty article.")
    print("They reveal what your provision prioritizes and defines.\n")
```

```python
#@title ## Step 3: Load Existing Treaty Excerpts

# These are real excerpts from existing instruments, shortened for comparison

treaty_excerpts = {
    "EU AI Act (Article 6)": 
        "AI systems that pose a risk of harm shall be subject to a mandatory risk assessment. "
        "Providers shall identify, analyse and evaluate the known and foreseeable risks associated with each AI system. "
        "High-risk AI systems shall implement appropriate risk mitigation measures.",
    
    "Council of Europe Framework Convention (Article 3)": 
        "Parties shall respect, protect and promote human rights and fundamental freedoms when developing, producing, "
        "procuring, deploying and using artificial intelligence systems, in accordance with their international and domestic legal obligations.",
    
    "OECD AI Principles": 
        "AI systems should be developed and deployed in a manner that respects human rights and fundamental freedoms. "
        "Stakeholders should support responsible stewardship of trustworthy AI. Transparency and explanations "
        "are critical to building and maintaining trust in AI systems.",
    
    "Global Digital Compact (AI Governance)": 
        "Establish multi-stakeholder approaches to ensure that artificial intelligence systems are designed, developed, "
        "deployed and used in a manner that respects human rights and the rule of law. Promote international cooperation "
        "on AI safety and the governance of AI systems."
}

print("="*80)
print("EXISTING TREATY EXCERPTS FOR COMPARISON")
print("="*80 + "\n")

for name, text in treaty_excerpts.items():
    print(f"\n{name}:")
    print(f"{text}\n")
    print("-" * 80)
```

```python
#@title ## Step 4: Embed and Compare

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

print("Loading embedding model (this may take a moment)...\n")
model = SentenceTransformer('all-MiniLM-L6-v2')

# Prepare texts for embedding
draft_summary = revised_draft.split('EXPLANATION')[0] if 'EXPLANATION' in revised_draft else revised_draft
texts_to_embed = [draft_summary[:1000]] + list(treaty_excerpts.values())
labels = ["Your Draft"] + list(treaty_excerpts.keys())

print("Generating embeddings...\n")
embeddings = model.encode(texts_to_embed)

# Compute similarity matrix
similarity_matrix = cosine_similarity(embeddings)

print("="*80)
print("SIMILARITY MATRIX: Your Draft vs. Existing Instruments")
print("="*80 + "\n")

# Compare your draft to each instrument
your_draft_similarities = similarity_matrix[0, 1:]

for i, (instrument, similarity) in enumerate(zip(treaty_excerpts.keys(), your_draft_similarities)):
    percentage = round(similarity * 100, 1)
    bar = "█" * int(percentage / 5) + "░" * (20 - int(percentage / 5))
    print(f"{instrument:45} {bar} {percentage}%")

print("\n" + "="*80)
most_similar_idx = np.argmax(your_draft_similarities)
most_similar_instrument = list(treaty_excerpts.keys())[most_similar_idx]
most_similar_score = your_draft_similarities[most_similar_idx]

print(f"\nCLOSEST MATCH: {most_similar_instrument} ({round(most_similar_score*100, 1)}% similar)")
print("\nINTERPRETATION: Your draft's 'genealogy' is closest to this instrument.")
print("This reflects your draft's language, concepts, and priorities.\n")
```

```python
#@title ## Step 5: Visualize in Treaty Space

# Use PCA to reduce embeddings to 2D for visualization
pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(embeddings)

# Create visualization
fig, ax = plt.subplots(figsize=(12, 8))

# Plot existing instruments
for i in range(1, len(labels)):
    ax.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1], 
               s=300, alpha=0.6, label=labels[i])
    ax.annotate(labels[i], (embeddings_2d[i, 0], embeddings_2d[i, 1]),
                xytext=(5, 5), textcoords='offset points', fontsize=10)

# Plot your draft (in bold)
ax.scatter(embeddings_2d[0, 0], embeddings_2d[0, 1],
           s=500, alpha=1.0, color='red', marker='*', label="Your Draft", edgecolors='darkred', linewidths=2)
ax.annotate("YOUR DRAFT", (embeddings_2d[0, 0], embeddings_2d[0, 1]),
            xytext=(5, -15), textcoords='offset points', fontsize=12, fontweight='bold', color='red')

ax.set_xlabel(f'PCA Dimension 1 ({pca.explained_variance_ratio_[0]:.1%} variance)', fontsize=11)
ax.set_ylabel(f'PCA Dimension 2 ({pca.explained_variance_ratio_[1]:.1%} variance)', fontsize=11)
ax.set_title('Your Draft in "Treaty Space": Relative Position to Existing Instruments\n(Closer = More Similar)', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='best', fontsize=10)

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("VISUALIZATION: Your Draft in Treaty Space")
print("="*80)
print(f"\nYour draft is positioned at ({embeddings_2d[0, 0]:.2f}, {embeddings_2d[0, 1]:.2f}) in this 2D space.")
print("\nThe visualization shows the conceptual landscape of AI governance instruments.")
print("Instruments that are close together share similar language and priorities.\n")
```

You learned that AI is not magic. It is **subsymbolic computation**—matrix multiplication at scale. Large language models like Gemini work through learned patterns, not explicit rules. This matters for law because:

- Speed: treaty drafting that would take weeks can happen in hours
